```python
import pandas as pd
import json

# Load your parquet file
df = pd.read_parquet("samples_3k_project_c_updated.parquet")

# Check dimensions and columns
print(df.shape)
df.info()

# Peek at first few rows
pd.set_option('display.max_colwidth', 120)
display(df.head(3))
```

```python
def safe_json(x):
    if isinstance(x, (dict, list)):
        return x
    if pd.isna(x):
        return {}
    try:
        return json.loads(x)
    except Exception:
        return {}

json_cols = [
    "names", "categories", "addresses", "websites", "socials", "emails", "phones",
    "base_names", "base_categories", "base_addresses", "base_websites", "base_socials",
    "base_emails", "base_phones"
]

for c in json_cols:
    if c in df.columns:
        df[c] = df[c].apply(safe_json)
```

```python
def get_primary_name(d):
    if isinstance(d, dict):
        return d.get("primary", "")
    return ""

def get_address_text(addr):
    if not addr:
        return ""
    if isinstance(addr, list):
        addr = addr[0] if len(addr) > 0 else {}
    return " | ".join(str(addr.get(k, "")) for k in ["freeform", "locality", "region", "country"] if addr.get(k))

# Left side
df["name"] = df["names"].apply(get_primary_name)
df["address"] = df["addresses"].apply(get_address_text)
df["website"] = df["websites"].apply(lambda x: x[0] if isinstance(x, list) and len(x)>0 else "")
df["phone"] = df["phones"].apply(lambda x: x[0] if isinstance(x, list) and len(x)>0 else "")

# Base side
df["base_name"] = df["base_names"].apply(get_primary_name)
df["base_address"] = df["base_addresses"].apply(get_address_text)
df["base_website"] = df["base_websites"].apply(lambda x: x[0] if isinstance(x, list) and len(x)>0 else "")
df["base_phone"] = df["base_phones"].apply(lambda x: x[0] if isinstance(x, list) and len(x)>0 else "")
```

```python
def normalize_text(s):
    if not isinstance(s, str):
        return ""
    return " ".join(s.lower().strip().split())

cols_to_clean = ["name", "address", "website", "phone", "base_name", "base_address", "base_website", "base_phone"]
for c in cols_to_clean:
    df[c] = df[c].apply(normalize_text)

df = df.dropna(subset=["name", "base_name"])
print("Cleaned shape:", df.shape)
```

```python
pair_keys = df.apply(lambda r: tuple(sorted([r["name"], r["base_name"]])), axis=1)
df = df.loc[~pair_keys.duplicated()].copy()

print("after cleaning:", df.shape)
```

```python
df_clean = df[[
    "label","id", "base_id",
    "name", "address", "website", "phone",
    "base_name", "base_address", "base_website", "base_phone",
    "confidence", "base_confidence"
]]

df_clean.to_parquet("places_cleaned.parquet", index=False)
df_clean.head()

```

```python
df = pd.read_parquet("places_cleaned.parquet")  # or data/cleaned/places_cleaned.parquet

# Select text columns you expect to be complete
text_cols = [
    "name", "address", "website", "phone",
    "base_name", "base_address", "base_website", "base_phone"
]

# Compute % of missing or blank values for each column
missing_pct = (
    df[text_cols]
    .apply(lambda x: x.isna() | (x.astype(str).str.strip() == ""))
    .mean() * 100
).round(2)

# Overall missing percentage across all text fields
overall_missing = missing_pct.mean().round(2)

print("Missing percentage by column (%):")
print(missing_pct)
print(f"\nAverage missing percentage across text fields: {overall_missing}%")
```
